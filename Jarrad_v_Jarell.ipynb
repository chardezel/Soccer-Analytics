```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for Jarell Quansah
quansah_data = {
    "Category": ["Tackles", "Interceptions", "Blocks", "Clearances", "Aerials Won", "Passes Attempted", "Passes Completed"],
    "Stats": [1.93, 1.20, 1.15, 2.92, 3.12, 91.04, 89.7]
}

# Data for Jarrad Branthwaite
branthwaite_data = {
    "Category": ["Tackles", "Interceptions", "Blocks", "Clearances", "Aerials Won", "Passes Attempted", "Passes Completed"],
    "Stats": [1.91, 1.44, 1.47, 4.71, 2.72, 40.00, 79.8]
}

# Create DataFrames
df_quansah = pd.DataFrame(quansah_data)
df_branthwaite = pd.DataFrame(branthwaite_data)

# Function to plot pie charts
def plot_pie_chart(df, title):
    plt.figure(figsize=(10, 6))
    plt.pie(df['Stats'], labels=df['Category'], autopct='%1.1f%%', startangle=140)
    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

# Plot pie chart for Jarell Quansah
plot_pie_chart(df_quansah, "Jarell Quansah's Stats")

# Plot pie chart for Jarrad Branthwaite
plot_pie_chart(df_branthwaite, "Jarrad Branthwaite's Stats")
```
