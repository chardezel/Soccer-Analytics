Alexandre Lacazette vs. Roberto Firmino in the Premier League 2018-2019 Season

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv ('england-premier-league-players-2018-to-2019-stats.csv')
df.head()
```

```python
player1 = 'Alexandre Lacazette'
player2 = 'Roberto Firmino'
player1_data = df[df['full_name'] == player1]
player2_data = df[df['full_name'] == player2]

print(player1_data[['full_name', 'goals_overall']])
print(player2_data[['full_name', 'goals_overall']])
```

```python
import json

config_data = {
  "player1": "Alexandre Lacazette",
  "player2": "Roberto Firmino"
}

with open('config.json', 'w') as f:
  json.dump(config_data, f)

with open('config.json', 'r') as f:
  config = json.load(f)

player1 = config['player1']
player2 = config['player2']

player1_data = df[df['full_name'] == player1]
player2_data = df[df['full_name'] == player2]

if player1_data['position'].values[0] != player2_data['position'].values[0]:
  raise ValueError("Players do not play the same position.")

print(player1_data[['full_name', 'goals_overall', 'assists_overall', 'appearances_overall']])
print(player2_data[['full_name', 'goals_overall', 'assists_overall', 'appearances_overall']])
```

```python
import pandas as pd
import json

file_path = 'england-premier-league-players-2018-to-2019-stats.csv'
data = pd.read_csv(file_path)

config_path = 'config.json'
with open(config_path, 'r') as file:
    config = json.load(file)
player1 = config['player1']
player2 = config['player2']
player_names = [player1, player2]

players_data = data[data['full_name'].isin(player_names)]

if players_data.empty:
    print("No valid players found for comparison.")
else:
    comparison_df = players_data.set_index('full_name').drop(columns=['position', 'nationality'])

    print(comparison_df)

comparison_df.to_csv('player_comparison.csv')
```
